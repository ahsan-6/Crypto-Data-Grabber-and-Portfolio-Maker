```python
import requests
import pandas as pd
```

```python
res=requests.get('https://api.wazirx.com/api/v2/tickers')
```

```python
data=(res.json())
data['btcinr']['last']
```

```python
df=pd.DataFrame(data)
df
```

```python
df2=pd.read_excel("Crypto.xlsx",index_col='Ticker')
df2
```

```python
tickers= list(df2.index)
print(tickers)
for i in tickers:
    df2['Last'][i]=df[i]['last']
df2
```

```python
df2.to_excel("Crypto.xlsx")
```
